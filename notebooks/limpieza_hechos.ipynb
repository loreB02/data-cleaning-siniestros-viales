```python
import pandas as pd

df = pd.read_csv("../datasets/raw/siniestros_viales_hechos.csv", sep=",", on_bad_lines="skip")
```

Exploro los datos previo a su manipulación

```python
df.info()
```

```python
df.head(10)
```

Elimino columnas irrelevantes

```python
df = df.drop_duplicates()
df = df.drop(columns= ["geocodificacion_plana", "longitud_siniestro", "latitud_siniestro", "rango_horario", "direccion_normalizada_siniestro", "anio_siniestro", "mes_siniestro", "dia_siniestro"])
df
```

Sumo la totalidad de Nan por columna

```python
nan_por_columna = df.isna().sum()
nan_por_columna
```

Manejo de valores Nan

```python
df["comuna_siniestro"] = df["comuna_siniestro"].fillna("Desconocido")
df["tipo_de_via_siniestro"] = df["tipo_de_via_siniestro"].fillna("Desconocido")
```

Relleno los valores Nan de hora_siniestro con su Moda

```python
hora_frecuente = df["hora_siniestro"].mode()[0]
df["hora_siniestro"] = df["hora_siniestro"].fillna(hora_frecuente)
```

```python
df.isna().sum()
```

Normalizo las variables de texto

```python
cols_texto = [
    "id_siniestro",
    "tipo_de_via_siniestro",	
    "participantes_siniestro",
    "comuna_siniestro",
    "modo_desplazamiento_victima",
    "contraparte_siniestro",
    "gravedad_siniestro"
]

for col in cols_texto:
    df[col] = df[col].str.strip().str.lower()

df.head(10)
```

Transformación de fechas y horas

```python
df["fecha_siniestro"] = pd.to_datetime(df["fecha_siniestro"], errors="coerce")
df["dia"] = df["fecha_siniestro"].dt.day_name()
df["mes"] = df["fecha_siniestro"].dt.month_name()
df["año"] = df["fecha_siniestro"].dt.year
```

```python
df["hora_siniestro"] = pd.to_datetime(df["hora_siniestro"], format="%H:%M:%S", errors="coerce")
df["hora"] = df["hora_siniestro"].dt.hour   
```

Creación de Rango horario (mañana, tarde, noche, madrugada)

```python
def categorizar_hora(h):
    if 6 <= h <12:
        return "mañana"
    elif 12 <= h < 18:
        return "tarde"
    elif 18 <= h <= 24:
        return "noche"
    else:
        return "madrugada"
    
df["rango_horario"] = df["hora"].apply(categorizar_hora)
```

```python
df["rango_horario"].head(10) #probando la función
```

Creación de tipo día (laboral vs fin de semana)

```python
df["tipo_dia"] = df["dia"].apply(lambda x: "fin de semana" 
                                 if x in ["Saturday", "Sunday"] 
                                 else "laboral")
```
